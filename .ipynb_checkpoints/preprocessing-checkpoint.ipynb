```python
from sklearn import preprocessing
import numpy as np
import pandas as pd

f=open('F:\daima\迁移学习原\迁移学习\Domain Adaptation via Transfer Component Analysis\code\python\da_tool\preprocessing.csv')
sample=pd.read_csv(f)

sample['S_Y'][sample['S_Y'] == 51]  = 3   
sample['S_Y'][sample['S_Y'] == 52]  = 3
sample['S_Y'][sample['S_Y'] == 21]  = 1
sample['S_Y'][sample['S_Y'] == 22]  = 1
sample['S_Y'][sample['S_Y'] == 41]  = 2
sample['S_Y'][sample['S_Y'] == 43]  = 2
sample['S_Y'][sample['S_Y'] == 113]  = 4
sample['S_Y'][sample['S_Y'] == 123]  = 4
sample['S_Y'][sample['S_Y'] == 122]  = 4
sample['T_Y'][sample['T_Y'] == 51]  = 3   
sample['T_Y'][sample['T_Y'] == 52]  = 3
sample['T_Y'][sample['T_Y'] == 21]  = 1
sample['T_Y'][sample['T_Y'] == 22]  = 1
sample['T_Y'][sample['T_Y'] == 41]  = 2
sample['T_Y'][sample['T_Y'] == 43]  = 2
sample['T_Y'][sample['T_Y'] == 113]  = 4
sample['T_Y'][sample['T_Y'] == 123]  = 4
sample['T_Y'][sample['T_Y'] == 122]  = 4
sample=sample[sample['S_C1'] != 0]

Xs=sample.iloc[:,1:37]
Xs = Xs.values
Ys = sample.iloc[:,37:38]
Ys = Ys.values
Xt=sample.iloc[:,38:74]
Xt = Xt.values
Yt = sample.iloc[:,74:75]
Yt = Yt.values
CM = sample.iloc[:,75:76]
CM = CM.values
scaler = preprocessing.StandardScaler().fit(Xs)
Xs=scaler.transform(Xs)
Xt=scaler.transform(Xt)
S = np.hstack((Xs,Ys))
T = np.hstack((Xt,Yt))
A = np.hstack((S,T))
train_test= np.hstack((A,CM))
np.savetxt('pre_standard.csv', train_test, delimiter=',', fmt='%.6f')




```
